```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Apply the default theme
sns.set_theme()

# Load an example dataset
flight_data = sns.load_dataset("flights")

# Create a visualization
sns.relplot(
    data = flight_data,
    x="year", y="passengers"
)
plt.show()
```

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
# Import seaborn --- scatter plot
import seaborn as sns

# Apply the default theme
#sns.set_theme()
#sns.set_style('whitegrid')
sns.set_style('darkgrid')

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="sex", style="smoker", size="tip",
    row=None,
    col_wrap=None,
    row_order=None,
    col_order=None,
    palette=None,
    hue_order=None,
    hue_norm=None,
    sizes=None,
    size_order=None,
    size_norm=None,
    markers="<<",
    dashes=None,
    style_order=None,
    legend='auto',
    kind='scatter',
    height=5,
    aspect=1,
    facet_kws=None,
    units=None
)

# Legend ----> {"auto", "brief", "full", or False}
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

```python
#Plots data and a regression model and fits across a facet grid
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
# Distplot - Figure-level interface for drawing distribution plots onto a FacetGrid. 
#Used for visualising uni-variate and bi-variate distribution of data.

sns.displot(data=tips, x="total_bill", col="time", kde=True)

# - :func:`histplot` (with ``kind="hist"``; the default)
# - :func:`kdeplot` (with ``kind="kde"``)
# - :func:`ecdfplot` (with ``kind="ecdf"``; univariate-only)
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=False, kind = "hist")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kind = "kde")
```

```python
#Lineplot
flights = sns.load_dataset("flights")
flights.head()
```

```python
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
may_flights = flights.query("month == 'Mar'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
may_flights = flights.query("month in('Jan', 'Mar', 'May')")
sns.lineplot(data=may_flights, x="year", y="passengers", hue = "month")
```

```python
# Emperical
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
```

```python
# catplot (swarm plot)
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

```python
sns.catplot(data=tips, kind="strip", x="day", y="total_bill", hue="smoker")
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
```

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```
